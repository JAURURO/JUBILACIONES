#Librería

```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
import seaborn.objects as so
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
archivo = "/content/drive/MyDrive/Python_Materia_y_Ejercicios/Codificación de Dojo/Bases_de_Datos_para_Pandas/Effective labour market exit age.xlsx"
```

```python
df = pd.read_excel(archivo)
```

```python
df
```

```python
df["Average Age"].value_counts()
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
#Definir el tamaño del gráfico antes de trabajarlo
plt.figure(figsize=(12, 6))
# Primer Gráfico
sns.barplot(x="Country", y="Average Age", data=df, hue= "Gender", width=0.8, errorbar=None)
plt.xticks(rotation=90)
#legenda (ubicación)
plt.legend(loc="lower left")
#modificacion de tamaños
plt.xlabel('Country', fontsize=15);
plt.ylabel('Average Age', fontsize=15);
plt.title('Edad de Jubilación Promedio desde 1970, según país', fontsize=20)
# Mostrar el gráfico
plt.show()
```

```python
#grafico 2 (solo 2023)
df2020 = df.loc[df["Year"] == 2020]
```

```python
df2020
```

```python
#Definir el tamaño del gráfico antes de trabajarlo
plt.figure(figsize=(12, 6))
# Segundo Gráfico
sns.barplot(x="Country", y="Average Age", data=df2020, hue= "Gender", width=0.8, errorbar=None)
plt.xticks(rotation=90)
#legenda (ubicación)
plt.legend(loc="lower left")
#modificacion de tamaños
plt.xlabel('Country', fontsize=15);
plt.ylabel('Average Age', fontsize=15);
plt.title('Edad de Jubilación Promedio año 2020, según país', fontsize=20)
# Mostrar el gráfico
plt.show()
```
